## ETS Decomposition of Traffic Dataset:

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Read dataset

```python
traffic = pd.read_csv(('/content/traffic.csv'),
                       index_col =['DateTime'],
                       parse_dates = True)
```

## show dataset

```python
traffic.head()
```

## Apply ETS

```python
result = seasonal_decompose(traffic['Vehicles'], model ='multiplicative', period=24)
```

## Plot ETS

```python
result.plot()
plt.xticks(rotation=45)
plt.show()
```
