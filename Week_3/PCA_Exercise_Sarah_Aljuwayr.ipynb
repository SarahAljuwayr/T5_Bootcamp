# **PCA**







#Step 1 : import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

#Step 2: Importing the data set

###Import the dataset and distributing the dataset into X and y components for data analysis.




```python
wine = load_wine(as_frame=True)
df = wine.frame
df.shape
```

```python
# distributing into X and y
X = df.iloc[:, 0:13].values
y = df.iloc[:, 13].values
```

#Step 3 : Step 3: Splitting the dataset into the Training set and Test set





```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#Step 4: Feature Scaling
###Doing the pre-processing part on training and testing set such as fitting the Standard scale.






```python
scaler = preprocessing.StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
scaler
```

#Step 5: Applying PCA function
###Applying the PCA function into the training and testing set for analysis.




```python
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

#Step 6: Fitting Logistic Regression To the training set




```python
lr = LogisticRegression()
lr.fit(X_train_pca, y_train)
```

#Step 7: Predicting the test set result




```python
y_pred = lr.predict(X_test_pca)
```

#Step 8: Making the confusion matrix




```python
cm = confusion_matrix(y_test, y_pred)
```

#Step 9: Predicting the training set result




```python
y_pred_train = lr.predict(X_train_pca)
```

#Step 10: Visualizing the Test set results




```python
plt.figure(figsize=(8, 6))

# Plot actual results
plt.scatter(X_test_pca[y_test == 0, 0], X_test_pca[y_test == 0, 1], color='red', label='Class 0 (Actual)')
plt.scatter(X_test_pca[y_test == 1, 0], X_test_pca[y_test == 1, 1], color='blue', label='Class 1 (Actual)')
plt.scatter(X_test_pca[y_test == 2, 0], X_test_pca[y_test == 2, 1], color='green', label='Class 2 (Actual)')

# Plot predicted results
plt.scatter(X_test_pca[y_pred == 0, 0], X_test_pca[y_pred == 0, 1], marker='x', s=100, color='orange', label='Class 0 (Predicted)')
plt.scatter(X_test_pca[y_pred == 1, 0], X_test_pca[y_pred == 1, 1], marker='x', s=100, color='purple', label='Class 1 (Predicted)')
plt.scatter(X_test_pca[y_pred == 2, 0], X_test_pca[y_pred == 2, 1], marker='x', s=100, color='cyan', label='Class 2 (Predicted)')

plt.title('Test Set Results Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
