```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\ASUS\Desktop\LABS\W2\D4\Salary Data.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,columns=["Gender","Education Level","Job Title"],drop_first=True)
```

```python
df
```

```python
x = df[['Years of Experience']]
y=df['Salary']
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model= LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='b', label='Actual Data')
plt.plot(X_test, y_pred, color='r', label='Regression Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Linear Regression')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse= mean_squared_error(y_test,y_pred)
r2= r2_score(y_test,y_pred)
print(f"Mean Squared Error:{mse:.2f}")
print(f"R_squared:{r2:.2f}")
```
