### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
import seaborn as sns

sns.pairplot(df2)
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df2, x='TrafficIndexLive', y='Weather_conditions')
plt.title('Traffic_Index vs. Weather_Conditions')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(10, 6))
sns.histplot(df2['City'], kde=True, bins=20)
plt.title('Histogram of Cities')
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

# Bar Chart

```python
sns.barplot(data=df2, x='City', y='JamsCount', palette='muted')
plt.title('Jams Count in each city')
plt.xticks(rotation=90)
plt.show()
```

As shown in Bar Chart above, The most Jams Count is Dubai.

# Heatmap

```python
correlation_matrix= df2.corr() #correlation matrices
```

```python
sns.heatmap(correlation_matrix, annot=True, fmt='.1f')
plt.show()
```

This heatmap shows the relationships between Number of variables. 

# HexBin plot

```python
df2.plot.hexbin(x='TrafficIndexLive', y='TravelTimeHistoric', gridsize=15)
plt.title('Travel Time Historic with Traffic Index Live')
plt.show()
```

This shows relationships between TrafficIndexLive and TravelTimeHistoric.
